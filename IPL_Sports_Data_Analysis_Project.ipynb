```python
print("Total Matches:", matches.shape[0])

# List of seasons
print("Seasons:", matches['season'].unique())

# Count matches per season
season_count = matches['season'].value_counts()
print(season_count)
```

```python
# Count how many matches each team won
team_wins = matches['winner'].value_counts()

print(team_wins.head(5))
```

```python
# Convert total_runs column to NumPy array
runs_array = np.array(deliveries['total_runs'])

# Basic calculations
print("Total Runs:", np.sum(runs_array))
print("Average Runs per Ball:", np.mean(runs_array))
print("Maximum Runs in a Ball:", np.max(runs_array))
```

```python
# Balls where more than 4 runs were scored
big_runs = runs_array[runs_array > 4]

print("Total Big Run Balls:", big_runs.size)
```

```python
# matches played per season
season_count = matches['season'].value_counts().sort_index()

plt.figure(figsize=(14,5))
plt.bar(season_count.index, season_count.values, color="lightgreen")
plt.title("Matches Played Per Season")
plt.xlabel("Season")
plt.ylabel("Matches")
plt.show()
```

```python
# most successful teams(most wins)
team_wins = matches['winner'].value_counts()

plt.figure(figsize=(12,6))
sns.set_style("whitegrid")
ax = sns.barplot(x=team_wins.values, y=team_wins.index, color="pink",
    edgecolor="red" )
plt.title("Most Successful IPL Teams")
plt.xlabel("Number of Wins")
plt.ylabel("Team")

for container in ax.containers:
      ax.bar_label(container)
plt.show()
```

```python
# Count matches per season
season_matches = matches['season'].value_counts().sort_index()

# Plot
plt.figure(figsize=(14,6))
sns.set_style("whitegrid")

sns.scatterplot(x=season_matches.index, y=season_matches.values, s=250, color="purple", hue=season_matches, palette="deep", style=season_matches)

plt.title("Number of Matches Played Per Season")
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.show()
```

```python
# win margin(runs)
sns.set_style("whitegrid")

sns.displot(data=matches, x="result_margin", bins=30, kde=True, height=6, color="lightgreen")

plt.title("Distribution of Win Margin (Runs)")
plt.show()
```

```python
# most match wins by team
plt.figure(figsize=(12,8))

ax=sns.countplot(data=matches, y="winner", order=matches["winner"].value_counts().index, color="peachpuff", edgecolor="sienna")
plt.title("Total Matches Won by Each Team", fontsize=14, fontweight="bold")
plt.xlabel("Number of Wins")
plt.ylabel("Team")

for container in ax.containers:
      ax.bar_label(container)
plt.show()
```

```python
# Player of the Match Analysis
player_awards = matches['player_of_match'].value_counts().head(10)

plt.figure(figsize=(12,7))
sns.set_style("whitegrid")

sns.lineplot(x=player_awards.values, y=player_awards.index, marker="o", linewidth=4,color="green", markerfacecolor="red",
             linestyle="dashdot", markersize=11)

plt.title("Top 10 Player of the Match Winners", fontsize=15, fontweight="bold")
plt.xlabel("Number of Awards")
plt.ylabel("Player")
plt.show()
```

```python
# Toss Decision Analysis
toss_decision = matches['toss_decision'].value_counts()

plt.figure()
plt.pie(toss_decision, labels=toss_decision.index,autopct='%1.1f%%')
plt.title("Toss Decision Analysis")
plt.show()
```

```python
# Toss Winner vs Match Winner
matches['toss_match_win'] = matches['toss_winner'] == matches['winner']

plt.figure(figsize=(6,5))

ax = sns.countplot(data=matches, x="toss_match_win", palette="Set2", hue="toss_match_win")

plt.title("Toss Winner vs Match Winner", fontsize=14, fontweight="bold")
plt.xlabel("Did Toss Winner Win Match?")
plt.ylabel("Number of Matches")

# Add numbers on bars
for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
# Top 10 Run Scorers in IPL
top_batsmen = deliveries.groupby('batter')['batsman_runs'].sum()
top_batsmen = top_batsmen.sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
ax = sns.barplot(x=top_batsmen.values, y=top_batsmen.index, palette="colorblind", hue=top_batsmen)
plt.title("Top 10 Run Scorers in IPL")
plt.xlabel("Total Runs")
plt.ylabel("Batsman")
plt.show()
```

```python
# Top 10 Wicket Takers
wickets = deliveries[deliveries['dismissal_kind'].notnull()]

top_bowlers = wickets['bowler'].value_counts().head(10)

plt.figure(figsize=(10,6))
ax = sns.barplot(x=top_bowlers.values, y=top_bowlers.index, color="#B5EAD7", edgecolor="black")
plt.title("Top 10 Wicket Takers in IPL")
plt.xlabel("Wickets")
plt.ylabel("Bowler")

for container in ax.containers:
      ax.bar_label(container)
plt.show()
```

```python
# Runs Scored Per Over
runs_per_over = deliveries.groupby('over')['total_runs'].mean()

plt.figure(figsize=(10,7))
plt.plot(runs_per_over.index, runs_per_over.values, linewidth=4, linestyle="--", color="red")
plt.title("Average Runs Per Over")
plt.xlabel("Over")
plt.ylabel("Average Runs")
plt.show()
```

```python
# Venues with Most Matches
top_venues = matches['venue'].value_counts().head(10)

plt.figure(figsize=(8,6))
sns.barplot(x=top_venues.values, y=top_venues.index, palette="bright", hue=top_venues)
plt.title("Top 10 IPL Venues")
plt.xlabel("Matches Hosted")
plt.ylabel("Venue")
plt.show()
```

```python
# Team Performance While Batting First
bat_first = matches[matches['toss_decision'] == 'bat']
bat_first_wins = bat_first['winner'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.ecdfplot(bat_first, linewidth=6)
plt.title("Wins While Batting First")
plt.xlabel("Wins")
plt.ylabel("Team")
plt.show()
```

```python
# top 10 bowlers by best economy rate
# Count total balls bowled
balls_bowled = deliveries.groupby("bowler")["ball"].count()

# Total runs conceded
runs_conceded = deliveries.groupby("bowler")["total_runs"].sum()

# Calculate economy rate (runs per over)
economy = (runs_conceded / balls_bowled) * 6

# Create dataframe
eco_df = pd.DataFrame({
    "balls": balls_bowled,
    "economy": economy
})

# Filter bowlers with at least 500 balls
eco_df = eco_df[eco_df["balls"] >= 500]

# Top 10 best economy (lowest values)
top_economy = eco_df.sort_values("economy").head(10)

plt.figure(figsize=(12,7))

plt.barh(
    top_economy.index,
    top_economy["economy"],
    color="peachpuff",
    edgecolor="sienna",
    linewidth=2
)

plt.title("Top 10 Bowlers by Best Economy Rate (Min 500 Balls)",
          fontsize=14, fontweight="bold")
plt.xlabel("Economy Rate (Runs per Over)")
plt.ylabel("Bowler")
```
